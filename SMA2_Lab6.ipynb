```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt

# Connect to Google
pytrends = TrendReq(hl='en-US', tz=360)

# Build Payload
keywords = ['Python', 'Data Science']
pytrends.build_payload(keywords, timeframe='today 5-y', geo='US')

# Interest Over Time
interest_over_time = pytrends.interest_over_time()
print(interest_over_time.head())

# Plotting the data
plt.plot(interest_over_time.index, interest_over_time['Python'], label='Python')
plt.plot(interest_over_time.index, interest_over_time['Data Science'], label='Data Science')
plt.xlabel('Date')
plt.ylabel('Interest')
plt.title('Interest Over Time')
plt.legend()
plt.show()

```

```python
from pytrends.request import TrendReq
import pandas as pd

# Connect to Google
pytrends = TrendReq(hl='en-US', tz=360)

# Build Payload
keyword = 'Python'
pytrends.build_payload([keyword], timeframe='now 1-d', geo='US')

# Get 15-minute interval data
interval_data = pytrends.interest_over_time()
interval_data.index = interval_data.index.floor('15min')

# Aggregate the data to hourly
hourly_data = interval_data.resample('H').sum()

print(hourly_data)
```

```python
interest_by_region = pytrends.interest_by_region()
print(interest_by_region.head())
```

```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt

# Connect to Google
pytrends = TrendReq(hl='en-US', tz=360)

# Define the topics or queries of interest
topics = ['Python', 'Data Science', 'Machine Learning']

# Build Payload and fetch interest over time
pytrends.build_payload(topics, timeframe='today 5-y', geo='US')
interest_over_time = pytrends.interest_over_time()

# Visualize Interest Over Time
plt.plot(interest_over_time.index, interest_over_time)
plt.xlabel('Date')
plt.ylabel('Interest')
plt.title('Interest Over Time')
plt.legend(topics)
plt.show()
```

```python
related_queries = pytrends.related_queries()
print(related_queries)
```
